```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info() #Esta função mostra informações do meu DF
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
# A função nunique(valores distintos) vai contabilizar a partir do IDCOMPRA, então o o Idcanavenda é agrupado pela função groupyby
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo") #Com essa função REPLACE é possível fazer a troca de nome dos valores de cada coluna
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
#Fazendo a subtituição de espaço em branco por " _ "

vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_") # função que faz a substituição
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
#Aqui criamos uma local (vendas_sem_nulos) para armazenar o dropna, que basicamente exclui todas as linhas nulas do DF
```

```python
#subtituindo nulos
vendas["estado"].fillna("MS", inplace=True)
# portanto com a função fillna é possivel substituir todos os nulos da coluna estado pela sigla MS
# Obs: o inplace faz com que essa mudança vá direto para o DF, seri a mesma coisa que fazer:

#                      vendas["estado"] = vendas["estado"].fillna("MS")
```

```python
vendas.info()
# agora podemos verificar que o estado ja não consta mais com dados nulos, apenas a coluna Preço
```

```python
media_preco = vendas["Preço"].mean() # Neste comando criamos a variavel media_preco que receberá a média dos preços da coluna PREÇO

vendas["Preço"].fillna(media_preco, inplace=True) #Aqui é feito essa substituição de espaço nulo pelas medias de preço
#Obs: como media_preco foi/é uma variavel criada, dentro da função fillna ela não pode estar dentro de aspas ("").
```

```python
vendas.info()
```

```python
vendas.head()
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()

#Obs: Desvio padrão maior que a média quer dizer que tem uma dispersão grande entre os dados
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correta = vendas.query("Preço < Preço_com_frete")
# função query utilizada como filtro, seria o WHERE do BD
```

```python
vendas_correta.describe()
```

```python
vendas_correta.head()
```

```python
vendas_correta.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()

#Aqui foi feio um filtro onde está filtran Esporte e Lazer apenas no estado de SP
```

```python
vendas_correta.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()

# Aqui vamos agrupar Nome_Departamento, pelo seu id, utilizando o nunique para valores distintos (a quantidade que aparece em cada departamento),
# e o sort_values para forma decrescente. Reset_index() -> serve só para deixar mais bonito
```

```python
round(vendas_correta.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)

#O agg (agg — Agregue usando uma ou mais operações.) aplicado no Preço_com_frete utilizando a média
# A função round serve para acertar quantas casas decimais você gostaria de utilizar
```

```python
vendas_correta.groupby("Data").idcompra.nunique().sort_values(ascending = False).reset_index()

# Aqui podemos ter a resposta de quanto que foi vendido por cada data, em ordem decrescente.
```

```python
vendas_correta["mes"] = vendas_correta["Data"].dt.month
vendas_correta.head()
# Ao colocar vendas_correta["mes"] você cria uma nova coluna no DF
# Essa nova coluna "mes" recebe os valores de "Data" do DF, deixando apenas o mês, isso é possivel com as funções:
#dt -> Date time
# month -> mês
```

```python
vendas_correta.groupby("mes").idcompra.nunique().sort_values(ascending = False)
```

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correta.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = vendas_correta.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()

# FAMOSO LEFT JOIN DO SQL, de maneira simples é só chamar a função:
#        tabela1.merge.(tabela2, how="left", on="chave_tabeça")
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_bandeira_idade = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").reset_index(),2)
agg_bandeira_idade
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_bandeira_idade["bandeira"], agg_bandeira_idade["idade"], color="green")
plt.ylabel("Média de idade")
plt.title("Idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"],color="green" )
plt.ylabel("Renda média")
plt.title("Renda média por canal de vendas", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
vendas_por_data = vendas_correta.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(20,12))

plt.xlabel("Data da venda")
plt.ylabel("Quantidad de vendas")


plt.plot(vendas_por_data["Data"], vendas_por_data["idcompra"], color="green", linewidth="2");
plt.show()
```

```python
agg_departamento_preco = round(vendas_correta.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_departamento_preco
```

```python
fig = plt.figure(figsize=(50,15))
plt.bar(agg_departamento_preco["Nome_Departamento"], agg_departamento_preco["Preço_com_frete"], color="green")
plt.xlabel("Nome dos departamemtos")
plt.ylabel("Preços com frete")

plt.title("Preço médio dos departamentos", loc="center", fontsize=18);
```
